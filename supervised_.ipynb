TO GET THE DATASET FROM THE RENEWABLE RESOURCES

```python
country=pd.read_csv("/content/countries-table.csv")
country.fillna(0)
```

```python
del country['country']
country.head()
```

```python
x =country.drop('rank',axis=1)
y =country['rank']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1, test_size =0.4)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred_linear=reg.predict(x_test)
print(y_pred_linear)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print("The mean square error:",mean_squared_error(y_test,y_pred_linear))
print("The r2 score",r2_score(y_test,y_pred_linear))
```

```python
print('Intercept of the model:',reg.intercept_)
print('coefficient of the line:',reg.coef_)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test,y_pred_linear,color='c',linestyle='-')
plt.show()
```

```python
plt.plot(x_test,y_test, '+', label='Actual values')
plt.plot(x_test,y_pred_linear, label='Predicted values')
plt.xlabel('x_test')
plt.ylabel('y_test')
plt.legend()
plt.show()
```

the above graph shown the regression plot of the dataset

                     NAVIE BAYES

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
predicted = model.predict(x_test)
print('actual:',y_test)
print('predicted Value:',predicted)
```

```python
from sklearn.metrics import (accuracy_score,f1_score)
predicted=model.predict(x_test)
accuracy=accuracy_score(predicted,y_test)
f1=f1_score(predicted,y_test,average='weighted')
print('accuracy:',accuracy)
print('f1:',f1)
```

              DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree=dtree.fit(x,y)
dtree.fit(x_train,y_train)
```

```python
y_pred_decision=dtree.predict(x_test)
print(y_pred_decision)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score, ConfusionMatrixDisplay
tree_accuracy = accuracy_score(y_pred_decision,y_test)
tree_confusion = confusion_matrix(y_test,y_pred_decision)
tree_display = ConfusionMatrixDisplay(confusion_matrix = tree_confusion)
```

```python
print('Accuracy:',tree_accuracy)
print(classification_report(y_test,y_pred_decision))
tree_display.plot()
plt.show()
```
